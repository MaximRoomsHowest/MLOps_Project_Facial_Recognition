```python
from pathlib import Path
folders = [
    r"D:\School\Howest\MLOps\Project\EmotionsArchive\Data_1K\Angry",
    r"D:\School\Howest\MLOps\Project\EmotionsArchive\Data_1K\Fear",
    r"D:\School\Howest\MLOps\Project\EmotionsArchive\Data_1K\Happy",
    r"D:\School\Howest\MLOps\Project\EmotionsArchive\Data_1K\Sad",
    r"D:\School\Howest\MLOps\Project\EmotionsArchive\Data_1K\Surprise"
]
exts = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'}
for f in folders:
    p = Path(f)
    if not p.exists():
        print("MISSING:", p)
        continue
    files = [x for x in p.rglob('*') if x.is_file() and x.suffix.lower() in exts]
    print(p, "->", len(files), "images")
```
